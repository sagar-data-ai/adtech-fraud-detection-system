1. CLICKS DATASET

### Clicks Table Summary

The Clicks table represents the raw advertising traffic generated from mobile ad campaigns.  
Each row corresponds to a single ad click event when a user interacts with an advertisement displayed on a publisher platform.

This table forms the foundation of the analytics pipeline because every install and revenue event originates from a click.

**Columns Description**

- **click_id** – Unique identifier for each click event.
- **campaign_id** – Identifier for the advertising campaign responsible for generating the click.
- **publisher_id** – Identifier for the publisher or platform where the advertisement was displayed.
- **user_id** – Unique identifier representing the user who clicked on the advertisement.
- **device_type** – Type of device used by the user (e.g., Android, iOS).
- **country** – Geographic location of the user at the time of the click.
- **ip_address** – IP address of the user device, useful for detecting suspicious traffic or bot activity.
- **timestamp** – Date and time when the click occurred.
- **cost** – Advertising cost associated with the click.

The Clicks table is the largest table in the dataset and contains intentionally introduced data issues such as missing values, duplicate records, and abnormal patterns to simulate real-world advertising traffic.

# **2. Intsall Dataset**

### Installs Table Summary

The Installs table captures app installation events that occur after users click on advertisements.  
Each row represents a successful app install attributed to a specific ad click.

This table is used to measure the effectiveness of advertising campaigns by linking user installs to the clicks that generated them. It plays a crucial role in conversion analysis and campaign performance evaluation.

**Columns Description**

- **install_id** – Unique identifier for each installation event.
- **click_id** – Identifier of the click event that led to the installation. This column links the install to the Clicks table for attribution analysis.
- **install_timestamp** – The exact date and time when the application was installed by the user.

The installs table enables analysts to track user conversion from click to install and helps evaluate campaign efficiency, publisher performance, and user acquisition effectiveness.

# **3. Revenue Dataset**

### Revenue Table Summary

The Revenue table records monetization events generated by users after installing the application.  
Each row represents a revenue-generating action performed by a user, such as an in-app purchase or ad interaction.

This table is essential for evaluating the financial performance of advertising campaigns because it links user activity to revenue generation. By connecting this table with the Clicks and Installs tables, analysts can measure metrics such as campaign profitability, user lifetime value (LTV), and return on ad spend (ROAS).

**Columns Description**

- **user_id** – Unique identifier of the user who generated the revenue event.
- **revenue** – The monetary value generated from the user action.
- **event_timestamp** – The date and time when the revenue event occurred.

The Revenue table allows analysts to track how acquired users generate value over time and helps businesses understand which campaigns and publishers bring the most profitable users.
