```python
import numpy as np
import pandas as pd
```

```python
df1 = pd.read_parquet('clicks_cleaned.parquet')
df2 = pd.read_parquet('installs.parquet')
df3 = pd.read_parquet('revenue.parquet')
```

```python
df1.info()
df1.head()
```

```python
df1['timestamp'] = pd.to_datetime(df1['timestamp'], errors='coerce')
```

```python
df1['timestamp'].isna().sum()
```

```python
df1 = df1.dropna(subset=['timestamp'])
```

```python
df1['click_date'] = df1['timestamp'].dt.date
df1['click_hour'] = df1['timestamp'].dt.hour
df1['click_minute'] = df1['timestamp'].dt.minute
df1['day_of_week'] = df1['timestamp'].dt.day_name()
df1['week_number'] = df1['timestamp'].dt.isocalendar().week
df1['month'] = df1['timestamp'].dt.month
```

```python
df1['hour_bucket'] = df1['timestamp'].dt.floor('H')
df1['minute_bucket'] = df1['timestamp'].dt.floor('T')
```

```python
df1['click_date'] = pd.to_datetime(df1['click_date'])
```

```python
df1.info()
```

```python
df1.head()
```

```python
###############################################
```

```python
df1.to_csv("clicks_f1.csv", index=False)
```

```python
###############################################
```

```python
df2['install_timestamp'] = pd.to_datetime(df2['install_timestamp'], errors='coerce')
```

```python
df2['install_timestamp'].isna().sum()
```

```python
df2 = df2.dropna(subset=['install_timestamp'])
```

```python
df2['install_date'] = df2['install_timestamp'].dt.normalize()
df2['install_hour'] = df2['install_timestamp'].dt.hour
df2['install_week'] = df2['install_timestamp'].dt.isocalendar().week
```

```python
df3['event_timestamp'] = pd.to_datetime(df3['event_timestamp'], errors='coerce')
```

```python
df3['event_timestamp'].isna().sum()
```

```python
df3 = df3.dropna(subset=['event_timestamp'])
```

```python
df3['revenue_date'] = df3['event_timestamp'].dt.normalize()
df3['revenue_week'] = df3['event_timestamp'].dt.isocalendar().week
df3['revenue_month'] = df3['event_timestamp'].dt.month
```

```python
#######
## JOINING ##
```

```python
df4 = pd.read_csv('clicks_f1.csv')
df5 = pd.read_csv('installs_f1.csv')
df6 = pd.read_csv('revenue_f1.csv')
```

```python
df = df4.merge(df5, on='click_id', how='left')
```

```python
df['install_flag'] = df['install_timestamp'].notna().astype(int)
```

```python
df['timestamp'] = pd.to_datetime(df['timestamp'], errors='coerce')
df['install_timestamp'] = pd.to_datetime(df['install_timestamp'], errors='coerce')
```

```python
df[['timestamp','install_timestamp']].dtypes
```

```python
df['time_to_install'] = (df['install_timestamp'] - df['timestamp']).dt.total_seconds()
```

```python
df6_agg = df6.groupby('user_id')['revenue'].sum().reset_index()
```

```python
df = df.merge(df6_agg, on='user_id', how='left')
```

```python
df['revenue'] = df['revenue'].fillna(0)
```

```python
print(df.shape)
df.head()
```

```python
df.to_csv("final_dataset.csv", index=False)
```
